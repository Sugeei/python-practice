```python
import pandas as pd
import numpy as np
```

```python
# apply with positional arguments
def subtract_and_divide(x, sub, add, divide=1):
    return (x - sub + add) / divide
index = pd.date_range("1/1/2000", periods=8)
df = pd.DataFrame(np.random.randn(8, 3), index=index, columns=["A", "B", "C"])
df.apply(subtract_and_divide, args=(1.5,0,), divide=1)
```
